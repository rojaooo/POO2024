```python
import csv

class Talhao:

    num_talhoes = []

    def __init__(self, id_talhao, area, fazenda):
        self.id_talhao = id_talhao
        self.area = area
        self.fazenda = fazenda

    @staticmethod
    def set_talhao(lista):
        id_talhao = int(input("Digite o ID do talhão (deve ser único): "))
        for talhao in lista:
            if talhao.id_talhao == id_talhao:
                print("ID já cadastrado. Por favor, escolha outro ID.")
                return None
        area = float(input("Digite a área do talhão (hectares): "))
        fazenda = str(input("Digite o nome da fazenda: ")).lower()
        talhao = Talhao(id_talhao, area, fazenda)
        lista.append(talhao)
        return talhao


    @staticmethod
    def salvar_talhoes_em_csv(lista, nome_arquivo):
        registros_existentes = set()

        # Lê os registros já existentes no arquivo para evitar duplicação
        try:
            with open(nome_arquivo, mode='r', newline='') as arquivo_csv:
                leitor = csv.reader(arquivo_csv)
                next(leitor, None)  # Pula o cabeçalho
                for linha in leitor:
                    registros_existentes.add((int(linha[0]), float(linha[1]), str(linha[2])))
        except FileNotFoundError:
            pass  # Se o arquivo não existir, apenas criaremos um novo

        # Adiciona apenas registros novos
        with open(nome_arquivo, mode='a', newline='') as arquivo_csv:
            writer = csv.writer(arquivo_csv)
            if not registros_existentes:
                writer.writerow(['ID', 'Área (hectares)', 'fazenda'])  # Adiciona cabeçalho se o arquivo estava vazio

            for talhao in lista:
                if (talhao.id_talhao, talhao.area, talhao.fazenda) not in registros_existentes:
                    writer.writerow([talhao.id_talhao, talhao.area, talhao.fazenda])


    def substituir_talhoes(lista, nome_arquivo):

        lista.clear()
        talhao = Talhao.set_talhao(lista)

        with open(nome_arquivo, mode='w', newline='') as arquivo_csv:
            writer = csv.writer(arquivo_csv)
            writer.writerow(['ID', 'Área (hectares)', 'fazenda'])

            for talhao in lista:
                writer.writerow([talhao.id_talhao, talhao.area, talhao.fazenda])
```

```python
import csv

class Praga:

    def __init__(self, nome, qnt, data, id_talhao, densidade):
        self.nome = nome
        self.qnt = qnt
        self.data = data
        self.id_talhao = id_talhao
        self.densidade = densidade

    @staticmethod
    def set_praga(lista, nome_arquivo_talhoes):
        nome = str(input("Digite o nome da praga: "))
        qnt = int(input("Digite a quantidade: "))
        data = str(input("Digite a data: ")).strip()
        id_talhao = int(input("Digite o ID do talhão: "))

        # Verifica se o talhão existe no arquivo
        talhao_existe = False
        try:
            with open(nome_arquivo_talhoes, mode='r', newline='') as arquivo_csv:
                leitor = csv.reader(arquivo_csv)
                next(leitor, None)  # Pula o cabeçalho
                for linha in leitor:
                    if int(linha[0]) == id_talhao:  # Verifica se o ID existe no arquivo
                        talhao_existe = True
                        area_talhao = float(linha[1])
                        break
        except FileNotFoundError:
            print(f"Erro: O arquivo '{nome_arquivo_talhoes}' não foi encontrado.")
            return None

        if not talhao_existe:
            print(f"Erro: Talhão com ID {id_talhao} não está cadastrado.")
            return None

        densidade = qnt / (area_talhao * 10000)
        praga = Praga(nome, qnt, data, id_talhao, densidade)
        lista.append(praga)
        return praga


    @staticmethod
    def salvar_pragas_em_csv(lista, nome_arquivo):

        registros_existentes = set()

        # Lê os registros já existentes no arquivo para evitar duplicação
        try:
            with open(nome_arquivo, mode='r', newline='') as arquivo_csv:
                leitor = csv.reader(arquivo_csv)
                next(leitor, None)  # Pula o cabeçalho
                for linha in leitor:
                    registros_existentes.add((str(linha[0]), int(linha[1]), str(linha[2]), int(linha[3]), float(linha[4])))
        except FileNotFoundError:
            pass

        with open(nome_arquivo, mode='a', newline='') as arquivo_csv:
            writer = csv.writer(arquivo_csv)
            if not registros_existentes:
                writer.writerow(['Nome', 'Quantidade', 'Data', 'ID (do talhão)', 'densidade (pragas/m²)'])  # Cabeçalho
            for praga in lista:
                if (praga.nome, praga.qnt, praga.data, praga.id_talhao, praga.densidade) not in registros_existentes:
                    writer.writerow([praga.nome, praga.qnt, praga.data, praga.id_talhao, praga.densidade])



    def substituir_pragas(lista, nome_arquivo, nome_arquivo_talhoes):

        lista.clear()
        praga = Praga.set_praga(lista, nome_arquivo_talhoes)

        with open(nome_arquivo, mode='w', newline='') as arquivo_csv:
            writer = csv.writer(arquivo_csv)
            writer.writerow(['Nome', 'Quantidade', 'Data', 'ID (do talhão)', 'densidade (pragas/m²)'])

            for praga in lista:
                writer.writerow([praga.nome, praga.qnt, praga.data, praga.id_talhao, praga.densidade])
```

```python
import csv

class Controle:

    def __init__(self, nome, antes, depois, data, id_talhao, eficiencia):
        self.nome = nome
        self.antes = antes
        self.depois = depois
        self.data = data
        self.id_talhao = id_talhao
        self.eficiencia = eficiencia


    @staticmethod
    def set_controle(lista, nome_arquivo_talhoes):
        nome = str(input("Digite o nome do controle: "))
        antes = int(input("Digite a quantidade de pragas antes da aplicação: "))
        depois = int(input("Digite a quantidade de pragas depois da aplicação: "))
        data = str(input("Digite a data: ")).strip()
        id_talhao = int(input("Digite o ID (do talhão): "))
        talhao_existe = False
        try:
            with open(nome_arquivo_talhoes, mode='r', newline='') as arquivo_csv:
                leitor = csv.reader(arquivo_csv)
                next(leitor, None)  # Pula o cabeçalho
                for linha in leitor:
                    if int(linha[0]) == id_talhao:  # Verifica se o ID existe no arquivo
                        talhao_existe = True
                        break
        except FileNotFoundError:
            print(f"Erro: O arquivo '{nome_arquivo_talhoes}' não foi encontrado.")
            return None

        if not talhao_existe:
            print(f"Erro: Talhão com ID {id_talhao} não está cadastrado.")
            return None

        eficiencia = (1 - (depois / antes)) * 100
        controle = Controle(nome, antes, depois, data, id_talhao, eficiencia)
        lista.append(controle)
        return controle


    def salvar_controles_em_csv(lista, nome_arquivo):

        registros_existentes = set()

        # Lê os registros já existentes no arquivo para evitar duplicação
        try:
            with open(nome_arquivo, mode='r', newline='') as arquivo_csv:
                leitor = csv.reader(arquivo_csv)
                next(leitor, None)  # Pula o cabeçalho
                for linha in leitor:
                    registros_existentes.add((str(linha[0]), int(linha[1]), int(linha[2]), str(linha[3]), int(linha[4]), float(linha[5])))
        except FileNotFoundError:
            pass

        with open(nome_arquivo, mode='a', newline='') as arquivo_csv:
            writer = csv.writer(arquivo_csv)
            if not registros_existentes:
                writer.writerow(['nome', 'antes', 'depois', 'data', 'ID (do talhão)', 'Eficiência (%)'])
            for controle in lista:
                if (controle.nome, controle.antes, controle.depois, controle.data, controle.id_talhao, controle.eficiencia) not in registros_existentes:
                    writer.writerow([controle.nome, controle.antes, controle.depois, controle.data, controle.id_talhao, controle.eficiencia])


    def substituir_controles(lista, nome_arquivo, nome_arquivo_talhoes):

        lista.clear()
        controle = Controle.set_controle(lista, nome_arquivo_talhoes)

        with open(nome_arquivo, mode='w', newline='') as arquivo_csv:
            writer = csv.writer(arquivo_csv)
            writer.writerow(['nome', 'antes', 'depois', 'data', 'ID (do talhão)', 'Eficiência (%)'])

            for controle in lista:
                writer.writerow([controle.nome, controle.antes, controle.depois, controle.data, controle.id_talhao, controle.eficiencia])
```

```python
if __name__ == "__main__":

    listaT = []
    listaP = []
    listaC = []

    while True:

        escolha = (input("""

Você deseja cadastrar um novo talhão? (t):
Você deseja substituir todos os talhões por um novo talhão? (st):
Você deseja consultar os talhões cadastrados? (ct):

Você deseja cadastrar uma nova praga? (p):
Você deseja substituir todas as pragas por uma nova praga? (sp):
Você deseja consultar as prags cadastradas? (cp):

Você deseja cadastrar um novo controle? (c):
Você deseja substituir todos os controles por um novo controle? (sc):
Você deseja consultar os controles cadastrados? (cc):

Você deseja sair? (s):

        """)).strip().lower()


        if escolha == 't':
            talhao = Talhao.set_talhao(listaT)
            if talhao:
                print(f"Novo talhão cadastrado!")

            Talhao.salvar_talhoes_em_csv(listaT, 'talhoes_cadastrados.csv')
            print("Lista de talhões salva em 'talhoes_cadastrados.csv'.")


        elif escolha == 'st':
            Talhao.substituir_talhoes(listaT, 'talhoes_cadastrados.csv')
            if talhao:
                    print(f"Todos os talhões substituídos! ")


        elif escolha == 'ct':
            with open('talhoes_cadastrados.csv', mode ='r',  newline='') as csvfile:
                spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
                for row in spamreader:
                    print(', '.join(row))



        elif escolha == 'p':
            praga = Praga.set_praga(listaP, 'talhoes_cadastrados.csv')
            if praga:
                print(f"Nova praga cadastrada!")

            Praga.salvar_pragas_em_csv(listaP, 'pragas_cadastradas.csv')
            print("Lista de pragas salva em 'pragas_cadastradas.csv'.")


        elif escolha == 'sp':
            Praga.substituir_pragas(listaP, 'pragas_cadastradas.csv', 'talhoes_cadastrados.csv')
            if praga:
                    print(f"Todas as pragas substituídas por!")


        elif escolha == 'cp':
            with open('pragas_cadastradas.csv', mode ='r',  newline='') as csvfile:
                spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
                for row in spamreader:
                    print(' '.join(row))



        elif escolha == 'c':
            controle = Controle.set_controle(listaC, 'talhoes_cadastrados.csv')
            if controle:
                print(f"Novo controle cadastrado!")

            Controle.salvar_controles_em_csv(listaC, 'controles_cadastrados.csv')
            print("Lista de controles salva em 'controles_cadastrados.csv'.")


        elif escolha == 'sc':
            Controle.substituir_controles(listaC, 'controles_cadastrados.csv', 'talhoes_cadastrados.csv')
            if controle:
                print(f"Todos os controles substituídos!")


        elif escolha == 'cc':
            with open('controles_cadastrados.csv', mode ='r',  newline='') as csvfile:
                spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
                for row in spamreader:
                    print(' '.join(row))


        elif escolha == 's':
            print("Saindo...")
            break

        else:
            print("Opção inválida. Por favor, escolha uma opção válida.")
```
